We set the 100/tau to compare the graph of (x,y) for this 3 method (Euler,velocity verlet, and Runge Kutta). In order to see which is most likely to become a one elipse for one period

we have the relationship N*tau = time, we set the time is 100 year to see the eclipse first

```python
def Euler(tau):
    t_Eu = [0]
    x_Eu = [1-0.967]
    y_Eu = [0]
    vx_Eu = [0]
    vy_Eu = [np.sqrt((0.00699*(1+0.967))/((1-0.967)))]
    r_Eu = [x_Eu[0]]
    E_Eu = []
    v_Eu = [vy_Eu[0]]


    for i in range(1,int(100/tau)):
        t_Eu.append(tau*i)
        x_Eu.append(x_Eu[i-1]+tau*vx_Eu[i-1])
        y_Eu.append(y_Eu[i-1]+tau*vy_Eu[i-1])
        r_Eu.append(((x_Eu[i]**2)+(y_Eu[i]**2))**(0.5))
        vx_Eu.append(vx_Eu[i-1]+tau*(-0.00699*x_Eu[i-1]/(r_Eu[i-1])**3))
        vy_Eu.append(vy_Eu[i-1]+tau*(-0.00699*y_Eu[i-1]/r_Eu[i-1]**3))

        v_Eu.append(np.sqrt(vx_Eu[i]**2+vy_Eu[i]**2))




    return x_Eu,y_Eu,r_Eu,vx_Eu,vy_Eu,v_Eu,t_Eu

```

Verlet method

```python
def Verlet(tau):
    t_vv = [0]
    x_vv = [1-0.967]
    y_vv = [0]
    vx_vv = [0]
    vy_vv = [np.sqrt((0.00699*(1+0.967))/((1-0.967)))]
    r_vv = [x_vv[0]]
    v_vv = [vy_vv[0]]

    for i in range(1,int(100/tau)):
        t_vv.append(tau*i)
        x_vv.append(x_vv[i-1]+tau*vx_vv[i-1]+0.5*np.power(tau,2)*(-0.00699*x_vv[i-1]/(r_vv[i-1])**3))
        y_vv.append(y_vv[i-1]+tau*vy_vv[i-1]+0.5*np.power(tau,2)*(-0.00699*y_vv[i-1]/(r_vv[i-1])**3))
        r_vv.append(((x_vv[i]**2)+(y_vv[i]**2))**(0.5))
        vx_vv.append(vx_vv[i-1]+0.5*tau*((-0.00699*x_vv[i-1]/(r_vv[i-1])**3)+(-0.00699*x_vv[i]/(r_vv[i])**3)))
        vy_vv.append(vy_vv[i-1]+0.5*tau*((-0.00699*y_vv[i-1]/(r_vv[i-1])**3)+(-0.00699*y_vv[i]/(r_vv[i])**3)))

        v_vv.append(np.sqrt(vx_vv[i]**2+vy_vv[i]**2))



    return x_vv,y_vv,v_vv,r_vv,t_vv
```

Runge-Kutta method

```python
def Runge(tau):
    t_rk = [0]
    x_rk = [1-0.967]
    y_rk = [0]
    vx_rk = [0]
    vy_rk = [np.sqrt((0.00699*(1+0.967))/((1-0.967)))]
    v_rk = [vy_rk[0]]
    r_rk = []
    r_rk1 = []
    r_rk2 = []
    r_rk3 = []

    for i in range(1,int(100/tau)):
        t_rk.append(tau*i)
        r_rk.append(np.sqrt((x_rk[i-1]**2)+(y_rk[i-1]**2)))

        kx1 = tau*vx_rk[i-1]
        lx1 = tau*(-0.00699*x_rk[i-1]/(r_rk[i-1])**3)
        ky1 = tau*vy_rk[i-1]
        ly1 = tau*(-0.00699*y_rk[i-1]/(r_rk[i-1])**3)
        r_rk1.append(np.sqrt(((x_rk[i-1]+0.5*kx1)**2)+((y_rk[i-1]+0.5*ky1)**2)))

        kx2 = tau*(vx_rk[i-1]+0.5*lx1)
        lx2 = tau*(-0.00699/(r_rk1[i-1]**3))*(x_rk[i-1]+0.5*kx1)
        ky2 = tau*(vy_rk[i-1]+0.5*ly1)
        ly2 = tau*(-0.00699/(r_rk1[i-1]**3))*(y_rk[i-1]+0.5*ky1)
        r_rk2.append(np.sqrt(((x_rk[i-1]+0.5*kx2)**2)+((y_rk[i-1]+0.5*ky2)**2)))

        kx3 = tau*(vx_rk[i-1]+0.5*lx2)
        lx3 = tau*(-0.00699/r_rk2[i-1]**3)*(x_rk[i-1]+0.5*kx2)
        ky3 = tau*(vy_rk[i-1]+0.5*ly2)
        ly3 = tau*(-0.00699/r_rk2[i-1]**3)*(y_rk[i-1]+0.5*ky2)
        r_rk3.append(np.sqrt(((x_rk[i-1]+kx3)**2)+((y_rk[i-1]+ky3)**2)))

        kx4 = tau*(vx_rk[i-1]+lx3)
        lx4 = tau*(-0.00699/r_rk3[i-1]**3)*(x_rk[i-1]+kx3)
        ky4 = tau*(vy_rk[i-1]+ly3)
        ly4 = tau*(-0.00699/r_rk3[i-1]**3)*(y_rk[i-1]+ky3)

        x_rk.append(x_rk[i-1]+(1/6)*(kx1+2*kx2+2*kx3+kx4))
        vx_rk.append(vx_rk[i-1]+(1/6)*(lx1+2*lx2+2*lx3+lx4))
        y_rk.append(y_rk[i-1]+(1/6)*(ky1+2*ky2+2*ky3+ky4))
        vy_rk.append(vy_rk[i-1]+(1/6)*(ly1+2*ly2+2*ly3+ly4))

        v_rk.append(np.sqrt(vx_rk[i]**2+vy_rk[i]**2))



    return x_rk,y_rk,vx_rk,vy_rk,r_rk,v_rk,t_rk
```

Q2 Kepler second law proof



---


dA/dt = L/2m

L is angular momentum = mv_thetar

r*v_theta = v_r

then we have,
dA/dt = 1/2*r*(v_r)

we want to show that for every year, the dA/dt is the same. Which satisfy the Kelper second law

we take two measurement on tau, which show that for a smaller tau, we have a higher accuracy for dA/dt




```python
def Kepler_sec(tau,a,b,c,d):
    change_of_area = []
    for i in range(int(1/tau),int(76/tau)):
            x_new = a[i]
            y_new = b[i]
            vx_new = c[i]
            vy_new = d[i]
            r = np.sqrt((x_new**2+y_new**2))
            v = np.sqrt(vx_new**2+vy_new**2)
            change_of_area.append(0.5*r*v)

    return change_of_area
```
